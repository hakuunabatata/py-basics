```python

from uuid import uuid4 as uuid

max = int(input('numero maximo:'))

no_sense_word =  str(uuid())

words = []

for x in range(max):
    word = ''

    for y in str(x ** 2 ):
        word += no_sense_word[int(y)]
    words.append(word)

print(words)
```
